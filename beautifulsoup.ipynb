```python
import requests
import bs4
```

```python
response = requests.get('https://www.forevermissed.com/amy-elizabeth-smith/#about')
print(response.text) # Print the output

```

```python
soup = bs4.BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
data = soup.findAll(attrs={'class':'txt user_text'})
for i in data:
    print(i.string)
```
